```python
data = data.dropna()
```

```python
data
```

```python
# Select price as target variable
y = data.iloc[:, -2].values
```

```python
# Select the fatures
x = data.iloc[:, [0,1,2,3,4]].values
```

```python
# Library to Split the train test set
from sklearn.model_selection import train_test_split as tts
```

```python
# Split the train test
x_train, x_test, y_train, y_test = tts(x, y, test_size=0.2, random_state=4)
```

```python
# Output is continuous data
# Choose linear regression model
from sklearn.linear_model import LinearRegression
```

```python
# Create the model
linreg = LinearRegression()
```

```python
# Fit the model
linreg.fit(x_train, y_train)
```

```python
# Predict the accuracy
linreg.score(x_test, y_test)
```
